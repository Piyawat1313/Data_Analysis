```python
import numpy as np
import matplotlib.pyplot as plt

 # Creating equally spaced 100 data in range 0 to 2*pi
theta = np.linspace(0, 2 * np.pi, 100)

 # Generating x and y data
xvals = 16 * ( np.sin(theta) ** 3 )+20
yvals = 13 * np.cos(theta) - 5* np.cos(2*theta) - 2 * np.cos(3*theta) - np.cos(4*theta)+17
X = np.row_stack((xvals, yvals))


plt.axis("equal")
plt.grid(True)
plt.plot(X[0,:],X[1,:],"r")
plt.show()
```

```python
# การเปลี่ยนขนาด
def tranform(A, X):
    return np.dot(A,X)
A = np.array([[0.5,0],[0,0.5]])
B = np.array([[2,0],[0,2]])
theta1 = np.linspace(0,2*np.pi, 100)
X_v = 16 * (np.sin(theta1) ** 3) + 20
Y_v = 13 * np.cos(theta) - 5 * np.cos(2 * theta1) - 2 * np.cos(3 * theta1) - np.cos(4 * theta1) + 17
X = np.row_stack((X_v, Y_v))
ax = tranform(A, X)
bx = tranform(B, X)
plt.axis("equal")
plt.grid(True)
plt.plot(X[0,:], X[1,:], "r")
plt.plot(ax[0,:], ax[1,:], "g")
plt.plot(bx[0,:], bx[1,:], "b")
plt.show()
```

```python
# หมุนตามเข็มนาฬิกา
def tranform(A, X):
    return np.dot(A,X)
theta1 = np.linspace(0,2*np.pi, 100)
X_v = 16 * (np.sin(theta1) ** 3) + 20
Y_v = 13 * np.cos(theta) - 5 * np.cos(2 * theta1) - 2 * np.cos(3 * theta1) - np.cos(4 * theta1) + 17
X = np.row_stack((X_v, Y_v))
angle_a = (np.pi / 6) * 6
A_matrix = np.array([
    [np.cos(angle_a), -np.sin(angle_a)],
    [np.sin(angle_a), np.cos(angle_a)]
])
res_a = tranform(A_matrix, X)

plt.plot(X[0,:], X[1,:],"r--" , label="Original")
plt.plot(res_a[0,:], res_a[1,:], "b--", label="Rotatd (pi)")
plt.axis("equal")
plt.grid(True)
plt.legend()
plt.show()
```

```python
#หมุนตามเข็มนาฬิกา จำนวน 6 ครั้ง
def rotation_matrix(theta):
    return np.array([
        [np.cos(theta), np.sin(theta)],
        [-np.sin(theta), np.cos(theta)]
    ])

def tranform(A, X):
    return np.dot(A, X)


theta4 = np.linspace(0, 2 * np.pi, 100)
X_v4 = 16 * (np.sin(theta4) ** 3) + 20
Y_v4 = 13 * np.cos(theta4) - 5 * np.cos(2 * theta4) - 2 * np.cos(3 * theta4) - np.cos(4 * theta4) + 17
X4 = np.row_stack((X_v4, Y_v4))
plt.axis("equal")
plt.grid(True)

plt.plot(X4[0, :], X4[1,:], label="Original")
for i in range(1, 7):
    angle = i * (np.pi / 6)
    R = rotation_matrix(angle)
    rotated = tranform(R, X4)
    plt.plot(rotated[0, :], rotated[1,:], label=f"Rotate {i} pi / 6")

plt.legend()
plt.show()
```

```python
# สะท้อนข้ามแกน ํ แล้วนำผลที่ได้ไปหมุนทวนเข็มนาฬิกา pi / 3
def tranform(A, X):
    return np.dot(A,X)

theta1 = np.linspace(0,2*np.pi, 100)
X_v = 16 * (np.sin(theta1) ** 3) + 20
Y_v = 13 * np.cos(theta) - 5 * np.cos(2 * theta1) - 2 * np.cos(3 * theta1) - np.cos(4 * theta1) + 17
X = np.row_stack((X_v, Y_v))

M2 = np.array([[-1, 0], [0, 1]])

R = np.array([
    [np.cos(np.pi / 3), -np.sin(np.pi / 3)],
    [np.sin(np.pi / 3), np.cos(np.pi / 3)]
])

# สะท้อน y ก่อนแล้วค่อยหมุน
step1 = tranform(M2, X)
result_a = tranform(R, step1)
plt.axis("equal")
plt.grid(True)
plt.plot(X[0,:], X[1,:], label = "Original")
plt.plot(step1[0,:], step1[1,:], label = "after")
plt.plot(result_a[0,:], result_a[1, :], label = "Rotated")
plt.legend()
plt.show()
```

```python
# หมุนก่อนแล้วค่อยสะท้อน
step_b = tranform(R, X)
result_b = tranform(M2, step_b)
plt.axis("equal")
plt.grid(True)
plt.plot(X[0,:], X[1,:], label = "Original")
plt.plot(step_b[0,:], step_b[1, :], label = "After")
plt.plot(result_b[0,:], result_b[1, :], label = "Rotated")
plt.legend()
plt.show()
```
