```python
import numpy as np
import sympy as sp
A = np.array([
    [1,2],
    [3,4],
    [5,6]
])
B = np.array([
    [7,8,9],
    [1,2,3],
    [4,5,6]
])
# การสร้างเมทริกแบบรวม เมทริก [A | B]
print(np.hstack((A, B)))
```

```python
A2 = sp.Matrix([
    [1,2],
    [3,4],
    [5,6]
])
B2 = sp.Matrix([
    [7,8,9],
    [1,2,3],
    [4,5,6]
])
#  การสร้างเมทริกแบบ [A | B]
aug = A2.row_join(B2)
sp.pprint(aug)
m , n = A2.shape # m = row, n = columns
sp.pprint(m)
sp.pprint(n)
```

```python
A4 = sp.Matrix([
    [1,2,-1,0],
    [2,4,-2,1],
    [3,6,0,-1]
])
row1, col1 = A4.rref()

# แสดงข้อมูลเป็น metrix ที่เป็น rref
sp.pprint(row1)

# col1 เป็นการบอกว่า เลข 1 อยู่ตำแหน่งไหนบ้าง
sp.pprint(col1)
```

```python
A5 = sp.Matrix([
    [2,1],
    [5,3]
])

# n เก็บขนาดของแถวว่ามีขนาดเท่าไร 
n = A5.rows

# สร้างเมทริกเอกลักษณ์
I = sp.eye(n)

# ทำการรวมเมทริกเอกลักษณ์เข้าไปที่ metrix ตัวใหม่
Metrix_A5_I = A5.row_join(I)

# สร้างเมทริกโดยทำให้เป็น rref
row2, col2 = Metrix_A5_I.rref()

# แบบไม่มีการ invert
left = row2[:, :n]

# ทำให้เป็น invert
inv = row2[:, n:]
sp.pprint(row2)

# เช็คว่า เมทริกฝั่งซ้าย || ฝั่งขวา เป็นเอกลักษณ์ หรือไม่
if inv == sp.eye(n):
    print("\n A is invertible and ")
    sp.pprint(inv)

# ถ้าไม่เป็นเอกลักษณ์ 
else: 

    # เอาไปคูณกับเมทริกตัวเดิมได้เลย
    sp.pprint(A5*inv)
```

```python
A6 = sp.Matrix([
    [1,2,3, 5],
    [4,5,6, 4],
    [7,8,9, 5],
    [10,11,12,4]
])
n2 = A6.rows
inter = sp.eye(n2)

MM = A6.row_join(inter)
row3, col3 = MM.rref()
left = row3[:, :n2]
inv2 = row3[:, n2:]
sp.pprint(row3)

# เช็คว่าผกผันหรือไม่
if  A6.det() != 0:
    print("Yes")
    sp.pprint(inv2)
else: 
    print("No Invert")
```

```python
A7 = sp.Matrix([
    [1,2,0],
    [3,4, -1]
])
row, col4 = A7.shape

# กาทำ metrix มูลฐาน
II = sp.eye(row)
agrument = A7.row_join(II)
R, pivots = agrument.rref()


B3 = R[:, :col4]
P = R[:, col4:]
sp.pprint(R)
print("\n")

print("B3: \n")
sp.pprint(B3)
print("\n")

print("P = \n")
sp.pprint(P)

print("PA7 == B3: ", (P*A7).equals(B3))
```
