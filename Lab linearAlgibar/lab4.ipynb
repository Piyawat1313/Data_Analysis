```python
def classify_system2(A, B, tol=1e-10): #Return one of: unique / infinite /
    Aug = A.row_join(B)
    rAug = Aug.rank()
    rC = A.rank()
    nvar = A.shape[1]
    if rC != rAug:
        return 'no solution', None
    if rC == nvar:
        x,y,z = symbols('x, y, z') 
        X = sp.linsolve((A, B), (x, y, z)) # or use A.LUsolve(B)
        return 'unique solution:', X
    x,y,z = symbols('x,y,z')
    X = sp.linsolve((A, B), (x, y, z)) # find general solutions for a system
    return 'infinitely many solutions', X



def classify_system4(A, B, tol=1e-10): #Return one of: unique / infinite /
    Aug = A.row_join(B)
    rAug = Aug.rank()
    rC = A.rank()
    nvar = A.shape[1]
    if rC != rAug:
        return 'no solution', None
    if rC == nvar:
        x,y,z,w = symbols('x, y, z, w') 
        X = sp.linsolve((A, B), (x, y, z, w)) # or use A.LUsolve(B)
        return 'unique solution:', X
    x,y,z,w = symbols('x,y,z,w')
    X = sp.linsolve((A, B), (x, y, z, w)) # find general solutions for a system
    return 'infinitely many solutions', X




def classify_system_6(A, B, tol=1e-10): #Return one of: unique / infinite /
    Aug = A.row_join(B)
    rAug = Aug.rank()
    rC = A.rank()
    nvar = A.shape[1]
    if rC != rAug:
        return 'no solution', None
    if rC == nvar:
        x,y = symbols('x, y') 
        X = sp.linsolve((A, B), (x, y)) # or use A.LUsolve(B)
        return 'unique solution:', X
    x,y = symbols('x,y')
    X = sp.linsolve((A, B), (x, y)) # find general solutions for a system
    return 'infinitely many solutions', X
```
