```python
import numpy as np
import sympy as sp

# span
def spaned(v1, v2, u):
    A = sp.Matrix.hstack(v1,v2)
    # TODO builed variable 
    rowJoint = A.row_join(u)
    
    if A.rank() == A.row_join(u).rank():
        pp = u,'is in the span of', A
        sp.pprint(pp)
        #TODO
        s1, s2 = sp.symbols('s1 s2')
        sol = sp.linsolve(rowJoint(s1, s2))
        sp.pprint(sol)
    else:
        pp = u,'is NOT in the span of', A
        sp.pprint(pp)
        sp.pprint(rowJoint.rref()[0])
```

```python
A1 = sp.Matrix([
    [1,2,3],
    [2,4,6],
    [-1,-2,-3]
])

rankA1 = A1.rank()
nullA1 = A1.nullspace()
nullityA1 = len(nullA1)
sp.pprint(rankA1)
sp.pprint(nullA1)
sp.pprint(nullityA1)
```

```python
var1 = sp.Matrix([1,0,1])
var2 = sp.Matrix([0,1,1])
var3 = sp.Matrix([1,1,2])

A3 = sp.Matrix.hstack(var1, var2, var3)
R3 = 3
if A3.rank() == A3.shape[1]:
    pp2 = A3,'the column vectors are linearly independent'
    sp.pprint(pp2)
    
    # ตรวจสอบว่าเป็น ฐานของ R3 หรือไม่
    if A3.shape[1] == R3:
        print("is Basis for R^3")
    else:
        print("No basis")
else: 
     pp2 = A3,'the column vectors are linearly DEPEDENT'
     print("Therefore they are not a Basis for R^3 \n")
     sp.pprint(pp2)
```
