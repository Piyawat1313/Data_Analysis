## 1 การแผ่ทั่ว (Span)
##### เพื่อความสะดวกในที่นี้จะมองเวกเตอร์ใน  $\mathbb{R}^n$ เป็นคอลัมน์ของเมทริกซ์ (ทั้งนี้จะมองเวกเตอร์เป็นแถวหรือคอลัมน์ของเมทริกซ์ก็ได้ผลลัพธ์ทำนองเดียวกัน)

### การตรวจสอบว่า $u \in span\{ v_1,v_2,\dots, v_k\}$ หรือไม่  คือ
 ##### -  การตรวจสอบว่ามี $s_1,s_2,\dots,s_k \in \mathbb{R}$ ซึ่ง $s_1v_1+s_2v_2+\dots+s_kv_k=u$ หรือไม่
 ##### - หรือตรวจสอบว่าระบบสมการ $A{S}=u$ มีผลเฉลยหรือไม่ เมื่อ $A=\begin{bmatrix} v_1&v_2&\cdots& v_k\end{bmatrix}$ และ $S=\begin{bmatrix} s_1 \\s_2\\ \vdots\\s_k\end{bmatrix}$
 ##### - หรือตรวจสอบว่าระบบสมการ $rank(A)=rank([A|u])$ หรือไม่ 
### สำหรับ $u \in span\{ v_1,v_2,\dots, v_k\}$ เราสามารถหา $s_1,s_2,\dots,s_k \in \mathbb{R}$ ซึ่ง $s_1v_1+s_2v_2+\dots+s_kv_k=u$ โดยการแก้ระบบสมการเชิงเส้น $A{S}=u$ 

```python
v1=sp.Matrix([1,0,1])
v2=sp.Matrix([0,1,1])
u=sp.Matrix([1,1,2])
A=sp.Matrix.hstack(v1,v2)
if A.rank()==A.row_join(u).rank():
    pp=u,'is in the span of', A
    sp.pprint(pp)
    # it is your duty to find s_1,s_2,\dots,s_k  such that  s_1v_1+s_2v_2+\dots+s_kv_k=u
else:
    pp=u,'is NOT in the span of', A
    sp.pprint(pp) 
```

### แบบฝึกหัด จงเติมโปรแกรมด้านบนให้สมบูรณ์ สามารถหา $s_1,s_2,\dots,s_k \in \mathbb{R}$ ซึ่ง $s_1v_1+s_2v_2+\dots+s_kv_k=u$ ในกรณีที่ $u \in span\{ v_1,v_2,\dots, v_k\}$ พร้อมตรวจสอบ แล้วทดสอบโปรแกรมด้วย  $u$ และ $ v_1,v_2,\dots, v_k\}$  ต่อไปนี้
#### -  $u=\begin{bmatrix} 6 \\ 5 \\ 8\\ 9\end{bmatrix}$ และ $v_1=\begin{bmatrix} 1  \\  0 \\ 1 \\ 0 \end{bmatrix}$, $v_2=\begin{bmatrix}  0 \\ 1  \\ 0 \\ 1  \end{bmatrix}$, $v_3=\begin{bmatrix}  -1 \\  0 \\  0 \\ 2 \end{bmatrix}$
#### -  $u=\begin{bmatrix} -1 \\ 0 \\ 5\\ 2\end{bmatrix}$ และ $v_1=\begin{bmatrix} 1  \\  0 \\ 1 \\ 0 \end{bmatrix}$, $v_2=\begin{bmatrix}  0 \\ 1  \\ 0 \\ 1  \end{bmatrix}$, $v_3=\begin{bmatrix}  -1 \\  0 \\  0 \\ 2 \end{bmatrix}$


## 2. เซตอิสระเชิงเส้น (Linear Independence Set)
###  การตรวจสอบว่าเซต $\{ v_1,v_2,\dots, v_k\}$ เป็นเซตอิสระเชิงเส้นหรือไม่  คือ
 ##### -  การตรวจสอบว่ามีเพียง $s_1=s_2=\dots=a_k=0$  ที่ทำให้  $s_1v_1+s_2v_2+\dots+s_kv_k=0$ หรือไม่
 ##### - หรือตรวจสอบว่าระบบสมการ $A{S}=0$ มีผลเฉลยเดียวหรือไม่ เมื่อ $A=\begin{bmatrix} v_1&v_2&\cdots& v_k\end{bmatrix}$ และ $S=\begin{bmatrix} s_1 \\s_2\\ \vdots\\s_k\end{bmatrix}$  
 ##### - หรือตรวจสอบว่า $rank(A)=$ จำนวนตัวแปร (จำนวนคอลัมน์ของ $A$) หรือไม่

```python
v1=sp.Matrix([1,2,0,-5])
v2=sp.Matrix([0,1,1,0])
v3=sp.Matrix([-2,1,1,3]) 
A=sp.Matrix.hstack(v1,v2,v3)
if A.rank() == A.shape[1]:
    pp=A, 'the colum vectors are linearly  independent'
    sp.pprint(pp)
else:
    pp=A, 'the colum vectors are linearly  DEPENSENT'     # not linearly independent
    sp.pprint(pp)
```

### แบบฝึกหัด จงตรวจสอบว่า  $ S=\{v_1,v_2,\dots, v_k\}$  ต่อไปนี้เป็นเซตอิสระเชิงเส้นหรือไม่
#### -  $S=\{v_1=\begin{bmatrix} 1  \\  0 \\ 1 \\ 0 \end{bmatrix}, v_2=\begin{bmatrix}  1 \\ 1  \\ 0 \\ 1  \end{bmatrix}, v_3=\begin{bmatrix}  5 \\  3 \\  2 \\ 3 \end{bmatrix} \}$
#### -  $S=\{v_1=\begin{bmatrix} 1  \\  0 \\ 1 \\ 0 \end{bmatrix}, v_2=\begin{bmatrix}  0 \\ 1  \\ 0 \\ 1  \end{bmatrix}, v_3=\begin{bmatrix}  -1 \\  0 \\  0 \\ 2 \end{bmatrix}, v_4=\begin{bmatrix} -1 \\ 0 \\ 5\\ 2\end{bmatrix}\}$

#### ข้อสังเกต ถ้า $ S=\{v_1,v_2,\dots, v_k\}$  เป็นเซตอิสระเชิงเส้น จะได้ว่า  $S$ เป็นฐาน(หลัก) ของ $span(S)$ เสมอ

## 3 ปริภูมิคอลัมน์และปริภูมิสู่ศูนย์ (Column Space and  Null Space)
#### สำหรับเมทริกซ์ $A$ ขนาด $m\times n$  จะได้ว่า $rank(A)=\dim(Col(A))$, (อาจใช้สัญลักษณ์) $nullity(A)=\dim(Null(A))$, และ $ rank(A)+  nullity(A)=n$

```python
A=sp.Matrix([[1,-3,-2,0],[-5,9,1,-2],[3,0,-3,2]])
colA=A.columnspace()
nullA=A.nullspace()
rankA=A.rank()
nullityA=len(nullA) # =A.shape[1]-rankA
pp='The column space of A is', colA
sp.pprint(pp)
pp='The rank of A is', rankA
sp.pprint(pp)
pp='The null space of A is', nullA
sp.pprint(pp)
pp='The nullity of A (or, dim(Null(A)) is', nullityA
sp.pprint(pp)
pp='The sum (nullity(A)+rank(A) is', nullityA+rankA
sp.pprint(pp)
```
