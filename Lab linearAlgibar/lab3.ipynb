```python
AA = np.array([
    [2,1,0,0],
    [0,3,4,0],
    [0,0,5,6],
    [0,0,0,7]
])
print(np.linalg.det(AA))
new_A = AA.copy()
new_A[2,3] = 10
print("new_A: ", np.linalg.det(new_A))
```

```python
metrix_A = np.array([
    [3,1,-1,4],
    [2,0,4,0],
    [1,0.5,3,6],
    [0,-3,0,2]
])
B = metrix_A.copy()
B[[0,1], :] = B[[1,0], :]
print("metrix_A: ",np.linalg.det(metrix_A))
print("metrix_B: ", np.linalg.det(B))
B2 = metrix_A.copy()
B2[[0, 1], :] = B2[[1,0], :]
B2[[2, 3], :] = B2[[3,2], :]
print(np.linalg.det(metrix_A))
print(np.linalg.det(B2))

B = metrix_A.copy()
B[2,:] = -2*B[2,:]
print(np.linalg.det(metrix_A))
print(np.linalg.det(B))

B[3, :] = B[3, :] + 5 * B[0, :]
print(np.linalg.det(metrix_A))
print(np.linalg.det(B))
```

```python
metrix2 = np.array([
    [3,1,-1,4],
    [2,0,4,0],
    [1,0.5,3,6],
    [0,-3,0,2]
])
swap = metrix2.copy()
swap[:, [0,1]] = swap[ :,[1,0]]
print(np.linalg.det(metrix2))
print(np.linalg.det(swap))

swap[:, [0,2]] = swap[:,[2, 0]]
print(np.linalg.det(swap))
swap[:, [2, 3]] = swap[:, [3, 2]]
print(np.linalg.det(swap))
swap[:, 1] = 5 * swap[:, 1]
print(np.linalg.det(swap))
swap[:, 1] = swap[:, 1] - 3 * swap[:, 0]
print(np.linalg.det(swap))
```

```python
Aw = np.array([
    [2,1],
    [5,3]
])
print(np.linalg.det(Aw))
if abs(np.linalg.det(Aw)) < 1e-10:
    print("A is not invert")
else:
    Ainv = np.linalg.inv(Aw)
    print(Ainv)
    print(Aw @ Ainv)
    
```

```python
As = sp.Matrix([
    [2,1],
    [5,3]
])
n = As.rows
I = sp.eye(n)

Aug = As.row_join(I)
R, pivot = Aug.rref()
left = R[:, :n]
right = R[:, n:]
sp.pprint(R)
if left == sp.eye(n):
    sp.pprint(right)
else:
    print(As*right)
```
