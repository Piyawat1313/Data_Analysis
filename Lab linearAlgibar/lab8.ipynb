```python
import numpy as np
import sympy as sp
from numpy import linalg as LA
from sympy import *
```

```python
# ตรวจสอบ vecter ลักษณะเฉพาะที่เป็นอิสระเชิงเส้นต่อกัน
A1 = sp.Matrix([[2,1],[1,2]])
A1.eigenvects()
```

```python
# ตรวจสอบสมการ Ax = lamda X
for eigval, mult, vects in A1.eigenvects():
    v = vects[0]
    print(eigval, A1 * v, eigval * v, sp.simplify(A1*v - eigval*v))
```

```python
A6 = sp.Matrix([
    [1,0,0,0,2],
    [1,1,0,0,0],
    [0,0,3,0,0],
    [2,0,-2,1,1],
    [0,2,0,0,-2]
])
for eigval, mult, vects in A6.eigenvects():
    v = vects[0]
    print(eigval, A6.eigenvals())
```

```python
eigenvals_dict = A6.eigenvals()
product = 1
for val, multiplicity in eigenvals_dict.items():
    product *= (val ** multiplicity)
    print(product)
```

```python
dets = A6.det()
if product == dets:
    print("equals")
else:
    print("not equals")
```

```python
A7 = np.array([
    [1,0,0,0,2],
    [1,1,0,0,0],
    [0,0,3,0,0],
    [2,0,-2,1,1],
    [0,2,0,0,-2]
])
eig_val, eig_vec = np.linalg.eig(A7)
for i in range(len(eig_val)):
    lam = eig_val[i]
    v = eig_vec[:, i]
    print(lam)
```

```python
for i in range(len(eig_val)):
    lam = eig_val[i]
    v = eig_vec[:, i]
    Avi = np.dot(A7, v)
    lam_vi = lam * v
    print(Avi)
is_equals = np.allclose(Avi, lam_vi)
if is_equals :
    print("equals")
else: 
    print("not equals")
```

```python
A_3x3 = np.random.randint(0, 9, (3,3))
vals, Q = np.linalg.eig(A_3x3)
L = np.diag(vals)
Q_inv = np.linalg.inv(Q)
result = Q @ L @ Q_inv
print(np.round(L, 2), "\n")
print(np.round(Q, 2), "\n")
print(np.round(result.real, 0))
```

```python
for n_size in [4, 5]:
    A_gen = np.random.randint(0,10, size=(n_size, n_size)) # สุ่มเมทริกซ์
    
    vals_gen, Q_gen = np.linalg.eig(A_gen)
    L_gen = np.diag(vals_gen)
    Q_inv_gen = np.linalg.inv(Q_gen)
    
    Result_gen = Q_gen @ L_gen @ Q_inv_gen

print("Result: ", Result_gen)
```
