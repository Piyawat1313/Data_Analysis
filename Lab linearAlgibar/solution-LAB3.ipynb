## การคำนวณดีเทอร์มิแนนต์ด้วย numpy

```python
import numpy as np

A = np.array([[3,1],[5,2]])
print("det(A) =", np.linalg.det(A))
```

### เฉลยแบบฝึกหัด 3.1

```python
# ข้อ 1
A1 = np.array([[3,1],[5,2]])
print("A1=",A1) 
print("det (A1)=", np.linalg.det(A1))

# ข้อ 2
A2 = np.array([[1,2,3],[2,4,6],[-5,0,0.5]])
print("\nA2=",A2) 
print("det (A2)=", np.linalg.det(A2))

# ข้อ 3
A3 = np.array([[1,0,2],[-1,3,4],[2,1,5]])
print("\nA3=",A3) 
print("det(A3) =", np.linalg.det(A3))

# ข้อ 4: random 6x6
A4 = np.random.randint(0,10, (6,6))
print("\nA4=",A4) 
print("det (A4)=", np.linalg.det(A4))

# ข้อ 5
A5 = np.array([[2,1,0,0],
               [0,3,4,0],
               [0,0,5,6],
               [0,0,0,7]])
print("\nA5=",A5) 
print("det(A) =", np.linalg.det(A5))

A5_1 = A5.copy()
A5_1[0,1] = 10
print("\nA5_1=",A5_1) 
print("det after changing a12 =", np.linalg.det(A5_1))

A5_2 = A5.copy()
A5_2[3,3] = 11
print("\nA5_2=",A5_2)
print("det after changing a44 =", np.linalg.det(A5_2))
```

## แบบฝึกหัด 3.2 ผลของการดำเนินการตามแถวต่อดีเทอร์มิแนนต์

```python
A = np.array([[3,1,-1,4],
              [2,0,4,0],
              [1,0.5,3,6],
              [0,-3,0,2]], dtype=float)

detA = np.linalg.det(A)
print("A =",A)
print("det(A)=",detA)

# swap row 1 and 2
B = A.copy()
B[[0,1],:] = B[[1,0],:]
print("\nB =",B)
print("det after swap =", np.linalg.det(B))

# scale row 3 by -2
B = A.copy()
B[2,:] *= -2
print("\nB =",B)
print("det after scaling row 3 =", np.linalg.det(B))

# R4 + 5R1 -> R4
B = A.copy()
B[3,:] += 5*B[0,:]
print("\nB =",B)
print("det after row add =", np.linalg.det(B))
```

##  แบบฝึกหัด 3.3 ผลของการดำเนินการตามคอลัมน์ต่อดีเทอร์มิแนนต์

```python
# swap columns 2 and 3
B = A.copy()
B[:,[1,2]] = B[:,[2,1]]
print("\nB =",B)
print("det after column swap =", np.linalg.det(B))

# swap columns 1 and 3// swap columns 3 and 4
B = A.copy()
B[:,[0,2]] = B[:,[2,0]]
B[:,[2,3]] = B[:,[3,2]]
print("\nB =",B)
print("det after column swap =", np.linalg.det(B))

# scale column 2 by 5
B = A.copy()
B[:,1] *= 5 #  B[:,1] = 5*B[:,1]
print("\nB =",B)
print("det after column scale =", np.linalg.det(B))

# C2 - 3C1 -> C2
B = A.copy()
B[:,1] = B[:,1] - 3*B[:,0]
print("\nB =",B)
print("det after column add =", np.linalg.det(B))
```

## แบบฝึกหัด 3.4 ดีเทอร์มิแนนต์และตัวผกผันของเมทริกซ์

```python
A = np.array([[2,1],[5,3]], dtype=float)
detA = np.linalg.det(A)
print("A =", A)
print("det(A) =", detA)

if abs(detA) < 1e-10:
    print("\nA is not invertible")
else:
    Ainv = np.linalg.inv(A)
    print("\nA^{-1} =\n", Ainv)
    print("\nA @ A^{-1} =\n", A @ Ainv)
    print("\nA*A^{-1} =I_2 (รวม error):", np.array_equal(A@Ainv,np.eye(2))) # มีการเก็บข้อผิดพลาดในตำแหน่งทศนิยมไกล ๆ
    print("\nA*A^{-1} =I_2:", np.allclose(A@Ainv,np.eye(2)))   #
```

```python
A = np.array([[2,3],[4,6]], dtype=float)
detA = np.linalg.det(A)
print("A =", A)
print("det(A) =", detA)

if abs(detA) < 1e-10:
    print("\nA is not invertible")
else:
    Ainv = np.linalg.inv(A)
    print("\nA^{-1} =\n", Ainv)
    print("\nA @ A^{-1} =\n", A @ Ainv)
    print("\nA*A^{-1} =I_2 (รวม error):", np.array_equal(A@Ainv,np.eye(2))) # มีการเก็บข้อผิดพลาดในตำแหน่งทศนิยมไกล ๆ
    print("\nA*A^{-1} =I_2:", np.allclose(A@Ainv,np.eye(2)))   #
```

## แบบฝึกหัด 3.5 ดีเทอร์มิแนนต์กับระบบสมการเชิงเส้น
  พิจารณาระบบสมการ
$
    	\begin{cases}
    		x+2y-z=3\\
    		2x+4y=6\\
    		3x-y+5z=1
    	\end{cases}
$

```python
A = np.array([[1,1,1],
              [2,-1,3],
              [3,2,-1]], dtype=float)
B = np.array([[6],[14],[5]], dtype=float)

print("A =", A)
print("B =", B)

detA = np.linalg.det(A)
print("det(A) =", detA)

if abs(detA) < 1e-10:
    print("No unique solution")
else:
    X = np.linalg.inv(A) @ B
    print("Solution X =", X)
```

## แบบฝึกหัด (พร้อมเฉลย)

```python
# Determinant powers
A = np.array([[1,-1,0],[5,0,-2],[-2,3,0.5]])
print("A =", A)
print("det(A) =", np.linalg.det(A))
print("det(A^2) =", np.linalg.det(A@A))
print("det(A^10) =", np.linalg.det(np.linalg.matrix_power(A,10)))

# Check invertibility
detA = np.linalg.det(A)
if abs(detA) < 1e-10:
    print("\nA not invertible")
else:
    Ainv=np.linalg.inv(A)
    print("\nA inverse =\n", Ainv)
    print("\nA * A^{-1} =\n", A @ Ainv)
    print("\nA*A^{-1} =I_3:", np.allclose(A@Ainv,np.eye(3)))   #

# Linear system
A = np.array([[2,1,-3],[1,-2,1],[1,-1,5]], dtype=float)
B = np.array([[1],[0],[2]], dtype=float)

print("A =", A)
print("B =", B) 

print("det(A) =", np.linalg.det(A))
if abs(np.linalg.det(A)) < 1e-10:
    print("\nNo unique solution")
else:
    print("\nSolution =", np.linalg.inv(A) @ B)

# Linear system
A = np.array([[2,1,-3],[1,-2,1],[3,-1,-2]], dtype=float)
B = np.array([[1],[4],[5]], dtype=float)

print("A =", A)
print("B =", B) 

print("det(A) =", np.linalg.det(A))
if abs(np.linalg.det(A)) < 1e-10:
    print("\nNo unique solution")
else:
    print("\nSolution =", np.linalg.inv(A) @ B)
```
