```python
from sympy import *
import sympy as sp
init_printing(use_unicode=True)
A = Matrix([
[4, 0, 1],
[2, 3, 2],
[1, 0, 4]
])
pprint(A)

U, S, V = A.singular_value_decomposition()
pprint(simplify(U))
pprint(simplify(S))
pprint(simplify(V))
print(A == simplify(U*S*(V.T)))
```

```python
A1 = sp.Matrix([
    [1,2,3],
    [4,5,6],
    [7,8,9]
])
U, S,V = A1.singular_value_decomposition()
sp.pprint(simplify(U))  #left
sp.pprint(simplify(S))  #singular Value เป็นเมทริกทแยงมุม
sp.pprint(simplify(V))  #right
sp.pprint(simplify(U * S * V.T))
```

```python
sp.pprint(simplify(U * U.T))  #U*U.T
sp.pprint(simplify(V.T * V))   #V.T * V
```

```python
# A * vi = si * ui
for i in range(S.rows):
    vi = V.col(i)
    ui = U.col(i)
    si = S[i, i]

    LHS = sp.simplify(A1 * vi)
    RHS = sp.simplify(A1 * ui)
    sp.pprint(LHS)
    sp.pprint(RHS)
    if((A1 * vi) == (si * ui)):
        print(True)
    else:
        print(False)
```

```python
# A.T * ui = si * vi
for i in range(S.rows):
    vi = V.col(i)
    ui = U.col(i)
    si = S[i, i]

    LHS = sp.simplify(A1 * vi)
    RHS = sp.simplify(A1 * ui)
    sp.pprint(LHS)
    sp.pprint(RHS)
    if((A1.T * ui) == (si * vi)):
        print(True)
    else:
        print(False)
```

```python
A2 = sp.randMatrix(4,5,0,9)
pprint(A2)
U, S,V = A2.evalf().singular_value_decomposition()
pprint(simplify(U))  #left
pprint(simplify(S))  #singular Value เป็นเมทริกทแยงมุม
pprint(simplify(V))  #right
pprint(simplify(U * S * V.T))
pprint(A == simplify(U*S*(V.T)))
```
