## 6.1 การแปลงเชิงเส้น

ให้ $V, W$ เป็นปริภูมิเวกเตอร์บน $\mathbb{R}$ เรียกฟังก์ชัน $T:V\to W$ ว่า **การแปลงเชิงเส้น** ก็ต่อเมื่อ

1. $T(u+v)=T(u)+T(v)$ สำหรับทุก $u,v\in V$  
2. $T(ku)=kT(u)$ สำหรับทุก $u\in V$ และ $k\in\mathbb{R}$

ในปฏิบัติการนี้พิจารณาเฉพาะกรณี $V=W\in\{\mathbb{R}^2,\mathbb{R}^3\}$ เท่านั้น


#### ทฤษฎีบท 6.1 (กรณี $ \mathbb{R}^2 \to \mathbb{R}^2$)

ถ้า $T:\mathbb{R}^2\to\mathbb{R}^2$ เป็นการแปลงเชิงเส้น แล้วมีจำนวนจริง $a,b,c,d$ ทำให้
$$
T(x,y)=(ax+by,cx+dy).
$$

เขียนในรูปเมทริกซ์ (แบบใน Lecture )ได้ว่า

$$
T\left(\begin{bmatrix}x\\y\end{bmatrix}\right)
=\begin{bmatrix}ax+by\\cx+dy\end{bmatrix}
= \begin{bmatrix}a&b\\c&d\end{bmatrix} 
\begin{bmatrix}x\\y\end{bmatrix}
= AX.
$$
โดยเรียก $A$ ว่า **เมทริกซ์การแปลง** ของ $T$.


#### ตัวอย่าง 6.1
ให้ $T:\mathbb{R}^2\to\mathbb{R}^2$ เป็นการแปลงเชิงเส้นนิยามโดย
$$
T(x,y)=(x+y,\,-y).
$$
เขียนในรูปเมทริกซ์ได้ว่า
$$
T\left(\begin{bmatrix}x\\y\end{bmatrix}\right)
=\begin{bmatrix}x+y\\-y\end{bmatrix}
= \begin{bmatrix}1&1\\0&-1\end{bmatrix} 
\begin{bmatrix}x\\y\end{bmatrix}
= AX.
$$

เมทริกซ์การแปลงคือ
$$
A=[T(e_1)~~T(e_2)]=\begin{bmatrix}1&1\\0&-1\end{bmatrix}.
$$
เขียนโปรแกรมเพื่อคำนวณ $T$ ของจุดใน $\mathbb{R}^2$ ดังนี้


````python
import numpy as np

def transform(A, X):
    """คำนวณการแปลงเชิงเส้น AX (A ขนาด 2x2, X ขนาด 2xn)"""
    return np.dot(A, X)

A = np.array([[1, 1],
              [0, -1]])

# จุด X (เขียนแบบคอลัมน์เวกเตอร์)
X = np.column_stack([[2, 3]])
ax = transform(A, X)

print(A, "maps\n", X, "to\n", ax)
````

คำนวณภาพของหลายจุดพร้อมกัน (เมทริกซ์จุดขนาด $2\times n$)

กำหนด $X$ เป็นเมทริกซ์ $2\times n$ ที่แต่ละคอลัมน์คือจุดใน $\mathbb{R}^2$ แล้วคำนวณ $AX$ ได้ทีเดียว


````python
import numpy as np

def transform(A, X):
    return np.dot(A, X)

A = np.array([[1, 1],
              [0, -1]])

# เมทริกซ์ 2x4 (4 จุด)
X = np.column_stack([[2, 3], [0, 1], [0, 5], [-1, 6]])
ax = transform(A, X)

print(A, "\nmaps\n", X, "\nto\n", ax)
````

##### การสร้างชุดข้อมูล $n$ จุดใน $\mathbb{R}^2$ (เมทริกซ์ขนาด $2\times n$)

การสร้างชุดข้อมูลด้วย `np.linspace` และ `np.column_stack` (ส่วนของเส้นตรง $y=0.5x$ จาก 0 ถึง 2)


````python
import numpy as np

def transform(A, X):
    return np.dot(A, X)

A = np.array([[1, 1],
              [0, -1]])

xvals = np.linspace(0, 2, 21)  # np.linspace(start, stop, num) เริ่ม จบ จำนวนค่า
X = np.column_stack([[x, 0.5*x] for x in xvals])  # เมทริกซ์ 2x21

ax = transform(A, X)
print(A, "\nmaps\n", X, "\nto\n", ax)
````

#### แบบฝึกหัด 6.1
ให้ $S:\mathbb{R}^2\to\mathbb{R}^2$ เป็นการแปลงเชิงเส้นนิยามโดย
$$
S(x,y)=(2x-y,\,-x+y).$$
จงเขียนโปรแกรมเพื่อคำนวณผลการแปลงของ
$$
X=
\begin{bmatrix}
1 & 3 & -1 & 0 & 2 & 0\\
0 & 1 & 2 & -1 & 0 & 0.5
\end{bmatrix}.
$$


````python
import numpy as np

def transform(A, X):
    return np.dot(A, X)

# เมทริกซ์การแปลงของ S(x,y) = (2x - y, -x + y)
A = np.array([[2, -1],
              [-1, 1]])

X = np.array([[1, 3, -1, 0, 2, 0],
              [0, 1,  2, -1, 0, 0.5]], dtype=float)

SX = transform(A, X)
print("X =\n", X)
print("\nS(X) =\n", SX)
````

## 6.2 การประกอบของการแปลงเชิงเส้น

ให้ $T:\mathbb{R}^2\to\mathbb{R}^2$ และ $S:\mathbb{R}^2\to\mathbb{R}^2$ เป็นการแปลงเชิงเส้น โดยมีเมทริกซ์การแปลงตามลำดับคือ $A$ และ $B$  

แล้วการประกอบ $S\circ T$ นิยามโดย $S(T(x,y))$ และมีเมทริกซ์การแปลงเป็น
\[
BA.
\]

#### ตัวอย่าง 6.2
คำนวณการประกอบ $S\circ T$ ด้วยการคูณเมทริกซ์ (หรือคำนวณทีละขั้น)


````python
import numpy as np

def transform(A, X):
    return np.dot(A, X)

A = np.array([[1, 1],
              [0, -1]])            # T
B = np.array([[2, -1],
              [-1, 1]])            # S

X = np.column_stack([[2, 3], [0, 1], [0, 5], [-1, 6]])

ax = transform(A, X)      # T(X)
bx = transform(B, ax)     # S(T(X))

# หรือทำครั้งเดียวด้วย BA
bx2 = transform(np.dot(B, A), X)

print("S∘T maps\n", X, "\nto\n", bx)
print("\nCheck with BA gives\n", bx2)
````

T(x,y)=(x+y,\,-y),\qquad S(x,y)=(2x-y,\,-x+y).

\[

\begin{bmatrix}

0 & 1 & 2 & -1 & 0 & 0.5

## 6.3 การแสดงผลการแปลงเชิงเส้นเชิงภาพสองมิติ (ใช้ matplotlib)

#### ตัวอย่าง 6.3
แสดงภาพของจุด/ภาพ ในระนาบเดียวกัน


````python
import numpy as np
import matplotlib.pyplot as plt

def transform(A, X):
    return np.dot(A, X)

A = np.array([[1, 0],
              [0, -1]])

X = np.column_stack([[2, 3]])
ax = transform(A, X)

plt.axis("equal")
plt.grid(True)
plt.plot(X[0, :], X[1, :], "o")
plt.plot(ax[0, :], ax[1, :], "o")
plt.show()
````

````python
import numpy as np

def transform(A, X):
    return np.dot(A, X)

A_T = np.array([[1, 1],
                [0, -1]])       # T
A_S = np.array([[2, -1],
                [-1, 1]])       # S

X = np.array([[1, 3, -1, 0, 2, 0],
              [0, 1,  2, -1, 0, 0.5]], dtype=float)

# 1) S∘T มีเมทริกซ์ A_S A_T
A_ST = A_S @ A_T
STX = transform(A_ST, X)

# 2) T^2 และ 3) T^5
A_T2 = np.linalg.matrix_power(A_T, 2)
A_T5 = np.linalg.matrix_power(A_T, 5)

T2X = transform(A_T2, X)
T5X = transform(A_T5, X)

print("S∘T (matrix)=\n", A_ST)
print("S∘T(X)=\n", STX)

print("\nT^2 (matrix)=\n", A_T2)
print("T^2(X)=\n", T2X)

print("\nT^5 (matrix)=\n", A_T5)
print("T^5(X)=\n", T5X)
````

#### ตัวอย่าง 6.5
แสดงภาพของข้อมูลทั่วไปด้วยสมการอิงตัวแปรเสริม


````python
import numpy as np
import matplotlib.pyplot as plt

def transform(A, X):
    return np.dot(A, X)

A = np.array([[1, 0],
              [0, -1]])

theta = np.linspace(0, 2*np.pi, 100)

xvals = 16*(np.sin(theta)**3) + 20
yvals = 13*np.cos(theta) - 5*np.cos(2*theta) - 2*np.cos(3*theta) - np.cos(4*theta) + 17

X = np.vstack((xvals, yvals))
ax = transform(A, X)

B= np.array([[1, 0.5],
              [-1, 0]])
bx = transform(B, X)


plt.axis("equal")
plt.grid(True)
plt.plot(X[0, :], X[1, :], "r", label="X")
plt.plot(ax[0, :], ax[1, :], "g", label="AX")
plt.plot(bx[0, :], bx[1, :], "b", label="BX")

plt.legend()  # คำอธิบายชื่อกราฟ ใช้คู่กับ label
plt.show()
````

#### แบบฝึกหัด 6.6 จากตัวอย่าง 6.3, 6.5, 6.7

1) ในแต่ละตัวอย่างให้เปลี่ยนเมทริกซ์ $A$ ตัวอย่างละ 3 แบบ (กำหนดเอง) และ plot ผลลัพธ์ในระนาบเดียวกัน  
2) ในแต่ละตัวอย่างให้ใช้เมทริกซ์
$$
A=\begin{bmatrix}0&-1\\1&0\end{bmatrix},\quad A^2,\quad A^3
$$
และ plot ผลลัพธ์ในระนาบเดียวกัน


````python
import numpy as np
import matplotlib.pyplot as plt

def transform(A, X):
    return np.dot(A, X)

# ตัวอย่างข้อมูล: ส่วนของเส้นตรง y = 0.5x
xvals = np.linspace(0, 2, 50)
X = np.column_stack([[x, 0.5*x] for x in xvals])

# (1) เปลี่ยน A 3 แบบ (กำหนดเอง)
As = [
    np.array([[1, 0],
              [0, 1]]),      # identity
    np.array([[2, 0],
              [0, 1]]),      # scale x
    np.array([[0, -1],
              [1,  0]])      # rotate 90 degrees
]

plt.axis("equal")
plt.grid(True)
plt.plot(X[0, :], X[1, :], "o", label="X")

for i, A in enumerate(As, 1):
    AX = transform(A, X)
    plt.plot(AX[0, :], AX[1, :], "o", label=f"A{i}X")

plt.legend()
plt.show()

# (2) ใช้ A, A^2, A^3
A = np.array([[0, -1],
              [1,  0]])

plt.axis("equal")
plt.grid(True)
plt.plot(X[0, :], X[1, :], "o", label="X")

for p in [1, 2, 3]:
    Ap = np.linalg.matrix_power(A, p)
    AX = transform(Ap, X)
    plt.plot(AX[0, :], AX[1, :], "o", label=f"A^{p}X")

plt.legend()
plt.show()
````

### ข้อ 2
จงเขียนโปรแกรมเพื่อคำนวณและแสดงผลการแปลงเชิงเส้นของข้อมูลในข้อ 1 โดยใช้เมทริกซ์
$$
A=\begin{bmatrix}1&0\\0&0\end{bmatrix},\quad
B=\begin{bmatrix}0.5&1\\0&2\end{bmatrix},\quad
BA,\quad AB
$$
พร้อมทั้ง plot ผลลัพธ์ในระนาบเดียวกัน


````python
import numpy as np
import matplotlib.pyplot as plt

def transform(M, X):
    return np.dot(M, X)

theta = np.linspace(0, np.pi/3, 100)
xvals = np.cos(theta) * np.sin(3*theta)
yvals = np.sin(theta) * np.sin(3*theta)
X = np.row_stack((xvals, yvals))

A = np.array([[1, 0],
              [0, 0]], dtype=float)

B = np.array([[0.5, 1],
              [0,   2]], dtype=float)

BA = B @ A
AB = A @ B

datasets = {
    "X": X,
    "AX": transform(A, X),
    "BX": transform(B, X),
    "BAX": transform(BA, X),
    "ABX": transform(AB, X),
}

plt.axis("equal")
plt.grid(True)
for name, Y in datasets.items():
    plt.plot(Y[0, :], Y[1, :], label=name)

plt.legend()
plt.show()

print("A=\n", A)
print("B=\n", B)
print("BA=\n", BA)
print("AB=\n", AB)
````

### ข้อ 3

````python
import numpy as np
import matplotlib.pyplot as plt

def transform(M, X):
    return np.dot(M, X)

theta = np.linspace(0, np.pi/3, 100)
xvals = np.cos(theta) * np.sin(3*theta)
yvals = np.sin(theta) * np.sin(3*theta)
X = np.vstack((xvals, yvals))

As = [
    np.array([[1, 0],
              [0, 1]], dtype=float),      # I
    np.array([[2, 0],
              [0, 0.5]], dtype=float),    # scale
    np.array([[0, 1],
              [1, 0]], dtype=float),      # swap
]

plt.axis("equal")
plt.grid(True)
plt.plot(X[0, :], X[1, :], label="X")

for i, A in enumerate(As, 1):
    Y = transform(A, X)
    plt.plot(Y[0, :], Y[1, :], label=f"A{i}X")

plt.legend()
plt.show()
````
