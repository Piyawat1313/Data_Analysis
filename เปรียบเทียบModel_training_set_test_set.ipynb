```python
import pandas as pd
import numpy as np
import sklearn
from sklearn.naive_bayes import GaussianNB, BernoulliNB 
from sklearn import preprocessing
from sklearn import metrics
from sklearn.neighbors import KNeighborsClassifier
print(np.__version__)
```

```python
data = pd.read_csv("Downloads/การเรียนมหาลัย/07-Lab-EXER-Compare-Performance-2Oct2025/Play_Golf.csv")
print(data)
```

```python
#แปลงข้อมูลเป็นตัวเลขทั้งหมด 
number = preprocessing.LabelEncoder()
data['outlook'] = number.fit_transform(data['outlook'])
data['temperature'] = number.fit_transform(data['temperature'])
data['humidity'] = number.fit_transform(data['humidity'])
data['windy'] = number.fit_transform(data['windy'])
print(data)
```

```python
#วิธีการทำ Traning set กับ Test set
from sklearn.model_selection import train_test_split
X = data[['outlook','temperature','humidity','windy']]
Y = data['play']
x_train, x_test, y_train, y_test = train_test_split(X,Y,test_size=0.3,random_state=5)
print(len(x_train),len(x_test),len(y_train),len(y_test))

X = data[['outlook','temperature','humidity','windy']]
Y = data['play']
x_train, x_test, y_train, y_test = train_test_split(X,Y,test_size=0.4,random_state=5)
print(len(x_train),len(x_test),len(y_train),len(y_test))
```

```python
#วิธีการฝึก model ของ GaussianNB
model_gau = GaussianNB()
model_gau.fit(data[['outlook','temperature','humidity','windy']],data['play'])
```

```python
#วิธีการทำให้ Model ทำนาย
model_gau.fit(x_train.values,y_train.values)
y_predic_gau = model_gau.predict(x_test.values)
print(model_gau.score(x_train.values, y_train.values))
print(model_gau.score(x_test.values, y_test.values))
print("Gaussian: ",metrics.accuracy_score(y_test,y_predic_gau))
```

```python
model_ber = BernoulliNB()
model_ber.fit(data[['outlook','temperature','humidity','windy']],data['play'])
```

```python
#วิธีการทำให้ Model ทำนาย
model_ber.fit(x_train.values, y_train.values)
y_predic_ber = model_ber.predict(x_test.values)
print(model_ber.score(x_train.values,y_train.values))
print(model_ber.score(x_test.values, y_test.values))
print("Bernoulli: ",metrics.accuracy_score(y_test,y_predic_ber))
```

```python
model_knn = KNeighborsClassifier(n_neighbors=3)
model_knn.fit(data[['outlook','temperature','humidity','windy']],data['play'])
```

```python
#วิธีการทำให้ Model ทำนาย
model_knn.fit(x_train.values,y_train.values)
y_predic_Knn = model_knn.predict(x_test.values)
print(model_knn.score(x_train.values, y_train.values))
print(model_knn.score(x_test.values, y_test.values))
print("KNN: ",metrics.accuracy_score(y_test,y_predic_Knn))
```
