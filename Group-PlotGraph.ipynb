## อ่านไฟล์ housing.csv

```python
#1.ใช้ matplotlib เพื่อสร้าง histogram ที่แสดงข้อมูล  SalePrice เป็น 10 ช่วง สีของ histogram เป็นสีเหลือง และขอบของ histogram เป็นแดง  
import numpy as np
import scipy as sp
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
ht = pd.read_csv("C:/Users/Guest-Student/Downloads/04-EXER-LAB-28Aug2025/housing.csv")
plt.hist(ht['SalePrice'],bins=10,color='yellow',edgecolor='red')
```

```python
#2.ใช้ seaborn เพื่อสร้าง scatter plot ระหว่าง GrLivArea, SalePrice  
sns.scatterplot(x='GrLivArea', y='SalePrice', data = ht)
```

```python
#3.สร้างกราฟวงกลมที่แสดงความสัมพันธ์ระหว่างลักษณะของบ้าน (HouseStyle) กับ ราคาขายบ้าน
ht.groupby(['HouseStyle'])['SalePrice'].count().plot(kind='pie', x='HouseStyle', y='SalePrice')
```

```python
#4.ใช้ seaborn เพื่อสร้าง histogram ที่แสดงข้อมูล  SalePrice เป็น 10 ช่วง สีของ histogram เป็นเขียว และขอบของ histogram สีม่วง  
sns.histplot(ht['SalePrice'],bins=10, color='green',edgecolor='purple',kde=True)
```

```python
#5.ใช้ seaborn เพื่อสร้าง linear regression plot ระหว่าง GrLivArea, SalePrice  
sns.regplot(x='GrLivArea',y='SalePrice',data=ht)
```

```python
#6 plot boxplot ของ LotArea
plt.boxplot(x='LotArea',data=ht)
```

```python
#7. plot กราฟแท่ง ระหว่าง SaleType และ SalePrice
sns.barplot(x='SaleType',y ='SalePrice', hue='YrSold', data=ht, estimator=np.mean)
```

```python
#8. plot กราฟแท่งเพื่อดูจำนวนของ LandContour กับ ราคาขาย SalePrice แยกสีตามปีที่ขาย 
sns.barplot(x='LandContour',y ='SalePrice', hue='YrSold', data=ht, estimator=np.mean)
```

```python
#9.plot กราฟเส้นแสดงความสัมพันธ์ระหว่าง MSSubClass และ Sa;ePrice โดยแสดงสีตามปีที่ขาย YrSold
sns.lineplot(x='MSSubClass',y ='SalePrice', hue='YrSold', data=ht, estimator=np.mean)
```

```python
#10. plot scatter plot plot ระหว่าง LotArea และ SalePrice โดยให้สีแสดงแยกตามปีที่ขาย YrSold 
sns.scatterplot(x='LotArea',y ='SalePrice', hue='YrSold', data=ht)
```
